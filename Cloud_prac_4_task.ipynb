**Imports**

```python
import requests
import pandas as pd
import ipywidgets as widgets
from google.colab.widgets import Grid
```

**Create a data frame from the dataset**

```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
pd.set_option("display.max_columns", None)
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

data_df = pd.DataFrame(data['result']['records'])

data_df.head()
```

**Get the year data for the cars**

```python
# Get earliest and latest year of car production
min_year = data_df['shnat_yitzur'].min()
max_year = data_df['shnat_yitzur'].max()

# Get number of years present in the dataset
number_of_years = max_year - min_year + 1

# Create array of 0 with the corresponding number of years
year_frequency_list = [0] * number_of_years

# For each index in the list, incremet the index for every car with the corresponding year
for index, row in data_df.iterrows():
  year_frequency_list[row["shnat_yitzur"] - min_year] += 1

print("Years: ")
for i in range(len(year_frequency_list)):
  print(i + min_year, ":", year_frequency_list[i], "produced ", end=", ")
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from plotly.offline import iplot
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
# Tab 1: Data Overview
tab1_content = widgets.Output()
with tab1_content:
    print("Data Overview:")
    display(data_df.describe())  # Summary statistics

# Tab 2: Raw Data
tab2_content = widgets.Output()
with tab2_content:
    print("Raw Data:")
    display(data_df)  # Full DataFrame
# Tab 3: Bar plot
tab3_content = widgets.Output()
with tab3_content:
    print("Charts:")
    plt.clf()  # Clear the current figure
    plt.figure(figsize=(20, 8))  # Increased figure size to accommodate labels

    # Create bar plot with labels
    left = [i for i in range(len(year_frequency_list))]
    height = year_frequency_list
    tick_labels = [(i + min_year) for i in range(len(year_frequency_list))]

    plt.bar(left, height, tick_label = tick_labels, width = 0.5)
    plt.show()  # Add this to display the plot
    plt.close()  # Close the figure to free memory


# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content,tab3_content])
tabs.set_title(0, 'Tab 1: Data Overview')
tabs.set_title(1, 'Tab 2: Raw Data')
tabs.set_title(2, 'Tab 3: Bar chart')


# Display Tabs
display(tabs)
```
